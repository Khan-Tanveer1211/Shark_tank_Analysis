# Dataframe  export 

For exporting data into mysql with convert all rows into a list of tuples

```python
y = [] 
for i in range(len(df2)):
    x =tuple(df2.iloc[i])
    y.append(x)
    
```

```python
file = open('shark_tank.txt','w')
for tuple in y:
    file.write(str(tuple) + ',' + '\n')
    
file.close()
    
```
